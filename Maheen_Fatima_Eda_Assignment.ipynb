```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel("data.xlsx")

```

```python
df
```

```python
# Pie Chart representing the number and percentage of each car type (car name)
car_type_counts = df['CarName'].value_counts()
plt.figure(figsize=(20, 20))
plt.pie(car_type_counts, labels=car_type_counts.index, autopct='%1.1f%%')
plt.title("Car Types")
plt.show()
```

```python
car_body_counts = df['carbody'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=car_body_counts.index, y=car_body_counts.values)
plt.title("Number of Cars by Car Body Type")
plt.xlabel("Car Body Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='enginesize', y='citympg')
plt.title("Engine Size vs City MPG")
plt.xlabel("Engine Size")
plt.ylabel("City MPG")
plt.show()
```

```python
correlation = df['stroke'].corr(df['highwaympg'])
print(f"Correlation between stroke and Highway MPG: {correlation}")
```

```python
X = df[['enginesize']]
y = df['price']

model = LinearRegression()
model.fit(X, y)

# Coefficients and intercept of the linear regression model
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")

# Model evaluation
r_squared = model.score(X, y)
print(f"R-squared: {r_squared}")

# Plotting the linear regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='red', linewidth=2)
plt.title("Linear Regression: Car Price vs Engine Size")
plt.xlabel("Engine Size")
plt.ylabel("Car Price")
plt.show()
```
